```python
#Function to scrape ONE page
titles=[]
authors=[]
urls=[]
def scrape1(soup):
    title_list = soup.find_all('a', {'class': 'bookTitle'})
    author_list = soup.find_all('span', {'itemprop': 'name'})
    url_list = soup.find_all('a', {'class': 'authorName'})
    for title in title_list:
        titles.append(title.get_text())
    for author in author_list:
        authors.append(author.get_text())


```

```python
#scrape the first page

driver.get('https://www.goodreads.com/shelf/show/multicultural-literature')
time.sleep(6)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

scrape1(soup)


    
```

```python
"""
Requires Wikipedia-API and Python 3
https://pypi.org/project/Wikipedia-API/
"""

import wikipediaapi

w = wikipediaapi.Wikipedia(language='en', extract_format=wikipediaapi.ExtractFormat.WIKI)


for name in authors:
    p = w.page(name)
    if p.exists():
        print(p.summary, '\n')
        print(p.fullurl, '\n')
        info.append(p.summary)

        fullurl.append(p.fullurl)

    else:
        info.append(name + ": No information available.\n")
        fullurl.append("No URL available")
c.writerow(info)
c.writerow(fullurl)
```
